```python
# Importing Necessary Libraries
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline
```

```python
# Read Dataset file

df = pd.read_csv('UniversalBank.csv')
df.head()
```

```python
# Checking for null values
df.isnull().sum()
```

```python
# Dropping ID and ZIP Code columns from the dataset
df1 = df.drop(["ID","ZIP Code"], axis = 1)
df1.head()
```

```python
plt.figure(figsize=(15,8))
plt.title("Correlation Heatmap", fontsize=20)
sns.heatmap(df1.corr(),annot = True, cmap='mako')
```

```python
zero_class = df1[df1.CreditCard==0]
zero_class.shape
```

```python
one_class = df1[df1.CreditCard==1]
one_class.shape
```

```python
# Income vs Experience scatter plot
plt.xlabel('Income')
plt.ylabel('Experience')
plt.scatter(zero_class['Income'],zero_class['Experience'], color = 'green', marker='+')
plt.scatter(one_class['Income'], one_class['Experience'], color = 'red', marker='.')
```

```python
# CCAvg vs Family scatter plot
plt.xlabel('CCAvg')
plt.ylabel('Family')
plt.scatter(zero_class['CCAvg'],zero_class['Family'], color = 'blue', marker='+')
plt.scatter(one_class['CCAvg'], one_class['Family'], color = 'red', marker='.')
```

```python
# Scaling the data using Standard Scaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled = scaler.fit(df1.drop('CreditCard',axis=1)).transform(df1.drop('CreditCard',axis=1))
df_scaled = pd.DataFrame(scaled, columns=df1.columns[:-1])
df_scaled.head()
```

```python
# Splitting the columns in to dependent variable (x) and independent variable (y).
x = df_scaled
y = df1['CreditCard']
```

# Implementation of SVM

```python
# Split data in to train and test 
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
# Apply SVM Model (Linear)
from sklearn.svm import LinearSVC, SVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

svc=SVC(gamma='auto') 

svc.fit(x_train, y_train)
y_pred=svc.predict(x_test)

print('Model accuracy (SVC) : {0:0.3f}'. format(accuracy_score(y_test, y_pred)))
```

```python
# Apply SVM Model
linear_svc = LinearSVC(dual='auto')

linear_svc.fit(x_train, y_train)
y_pred2=linear_svc.predict(x_test)

print('Model accuracy (Linear SVC) : {0:0.3f}'. format(accuracy_score(y_test, y_pred2)))
```

```python
# Confusion Matrix (SVC)
cm = confusion_matrix(y_test, y_pred)
cm_matrix = pd.DataFrame(data=cm, columns=['Actual Positive:1', 'Actual Negative:0'], 
                                 index=['Predict Positive:1', 'Predict Negative:0'])

sns.heatmap(cm_matrix, annot=True, fmt='d', cmap='mako')
```

```python
# Classification Report
print(classification_report(y_test,y_pred))
```

```python
# Confusion Matrix (Linear SVC)
cm = confusion_matrix(y_test, y_pred2)
cm_matrix = pd.DataFrame(data=cm, columns=['Actual Positive:1', 'Actual Negative:0'], 
                                 index=['Predict Positive:1', 'Predict Negative:0'])

sns.heatmap(cm_matrix, annot=True, fmt='d', cmap='mako')
```

```python
# Classification Report (Linear SVC)
print(classification_report(y_test,y_pred2))
```

```python
# Apply SVM model using different Kernel function 
linear=SVC(kernel='linear').fit(x_train,y_train)
rbf=SVC(kernel='rbf').fit(x_train,y_train)
poly=SVC(kernel='poly').fit(x_train,y_train)
sigmoid=SVC(kernel='sigmoid').fit(x_train,y_train)

# Accuracy check
y_pred1 = linear.predict(x_test)
print('Model accuracy with linear kernel : {0:0.3f}'. format(accuracy_score(y_test, y_pred1)))
y_pred2 = rbf.predict(x_test)
print('Model accuracy with Gaussian RBF kernel : {0:0.3f}'. format(accuracy_score(y_test, y_pred2)))
y_pred3 = poly.predict(x_test)
print('Model accuracy with Polynomial kernel : {0:0.3f}'. format(accuracy_score(y_test, y_pred3)))
y_pred4 = sigmoid.predict(x_test)
print('Model accuracy with Sigmoid kernel : {0:0.3f}'. format(accuracy_score(y_test, y_pred4)))
```

```python
# Confusion Matrix (Linear Kernel)
cm = confusion_matrix(y_test, y_pred1)
cm_matrix = pd.DataFrame(data=cm, columns=['Actual Positive:1', 'Actual Negative:0'], 
                                 index=['Predict Positive:1', 'Predict Negative:0'])

sns.heatmap(cm_matrix, annot=True, fmt='d', cmap='mako')
```

```python
# Classification Report (Linear Kernel)
print(classification_report(y_test,y_pred1))
```

```python
# Confusion Matrix (RBF Kernel)
cm = confusion_matrix(y_test, y_pred2)
cm_matrix = pd.DataFrame(data=cm, columns=['Actual Positive:1', 'Actual Negative:0'], 
                                 index=['Predict Positive:1', 'Predict Negative:0'])

sns.heatmap(cm_matrix, annot=True, fmt='d', cmap='mako')
```

```python
# Classification Report (RBF Kernel)
print(classification_report(y_test,y_pred2))
```

```python
# Confusion Matrix (Polynomial Kernel)
cm = confusion_matrix(y_test, y_pred3)
cm_matrix = pd.DataFrame(data=cm, columns=['Actual Positive:1', 'Actual Negative:0'], 
                                 index=['Predict Positive:1', 'Predict Negative:0'])

sns.heatmap(cm_matrix, annot=True, fmt='d', cmap='mako')
```

```python
# Classification Report (Polynomial Kernel)
print(classification_report(y_test,y_pred3))
```

```python
# Confusion Matrix (Sigmoid Kernel)
cm = confusion_matrix(y_test, y_pred4)
cm_matrix = pd.DataFrame(data=cm, columns=['Actual Positive:1', 'Actual Negative:0'], 
                                 index=['Predict Positive:1', 'Predict Negative:0'])

sns.heatmap(cm_matrix, annot=True, fmt='d', cmap='mako')
```

```python
# Classification Report (Sigmoid Kernel)
print(classification_report(y_test,y_pred4))
```
